```python
# Initial the map
import folium
await piplite.install(['folium'])

# Crear un mapa centrado en EE.UU.
site_map = folium.Map(location=[28.573, -80.649], zoom_start=5)

# Crear un objeto MarkerCluster para agrupar lanzamientos cercanos
marker_cluster = MarkerCluster().add_to(site_map)

# Agregar marcadores para cada lanzamiento
for index, row in spacex_df.iterrows():
    coordinate = [row["Lat"], row["Long"]]
    color = "green" if row["class"] == 1 else "red"  # Verde = éxito, rojo = fallo
    
    # Agregar marcador al mapa
    folium.Marker(
        location=coordinate,
        popup=f"Launch Site: {row['Launch Site']}, Outcome: {'Success' if row['class'] == 1 else 'Failure'}",
        icon=folium.Icon(color=color)
    ).add_to(marker_cluster)

# Mostrar el mapa con los lanzamientos agrupados
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # Aproximación del radio de la Tierra en km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance


# Coordenadas del sitio de lanzamiento (ejemplo con CCAFS LC-40)
launch_site_lat = 28.5634
launch_site_lon = -80.5768

# Coordenadas de puntos cercanos (Ejemplo: costa, carretera)
coast_lat, coast_lon = 28.5641, -80.5678
highway_lat, highway_lon = 28.5623, -80.5804

# Calcular distancias
distance_to_coast = calculate_distance(launch_site_lat, launch_site_lon, coast_lat, coast_lon)
distance_to_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Mostrar resultados
print(f"Distancia al mar: {distance_to_coast:.2f} km")
print(f"Distancia a la carretera: {distance_to_highway:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
await piplite.install(['folium'])
from folium import PolyLine

# Coordenadas del sitio de lanzamiento (ejemplo con CCAFS LC-40)
launch_site_lat, launch_site_lon = 28.5634, -80.5768

# Coordenadas de la ciudad más cercana (ajusta con las que obtengas en el mapa)
city_lat, city_lon = 28.6083, -80.8075  
railway_lat, railway_lon = 28.5721, -80.5855  
highway_lat, highway_lon = 28.5618, -80.5790  

# Calcular distancias
distance_to_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_to_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_to_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Agregar líneas conectando el sitio de lanzamiento con cada punto cercano
PolyLine([[launch_site_lat, launch_site_lon], [city_lat, city_lon]], color="red").add_to(site_map)
PolyLine([[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]], color="blue").add_to(site_map)
PolyLine([[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]], color="green").add_to(site_map)

# Agregar marcadores mostrando la distancia
for coord, distance, label in zip(
    [[city_lat, city_lon], [railway_lat, railway_lon], [highway_lat, highway_lon]],
    [distance_to_city, distance_to_railway, distance_to_highway],
    ["City", "Railway", "Highway"]
):
    folium.Marker(
        coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
        )
    ).add_to(site_map)

# Mostrar el mapa con las conexiones
site_map

```
